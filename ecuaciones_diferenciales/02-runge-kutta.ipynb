```python
from sympy import Function, cos, Symbol
from sympy.abc import w
```

```python
y_symbol = Symbol('y')
func = cos(2*w) - 4*y_symbol
func
```

```python
def f(w_v, y_v):
    return func.subs([(w, w_v), (y_symbol, y_v)]).evalf()
```

```python
alpha = 3
beta = 4
h = 0.01

y = {}
y[0] = beta

t = {}
t['f'] = beta
t[0] = alpha
```

```python
iter = int((t['f'] - t[0]) / h)
iter
```

```python
K1 = 0
K2 = 0
K3 = 0
K4 = 0
for i in range(1, iter + 1):
    K1 = h * f(alpha, beta)
    K2 = h * f(alpha + h/2, beta + K1/2)
    K3 = h * f(alpha + h/2, beta + K2/2)
    K4 = h * f(alpha + h, beta + K3)

    t[i] = t[0] + i*h
    y[i] = y[i - 1] + (1/6) * (K1 + 2*K2 + 2*K3 + K4)
    
    print(K1, K2, K3, K4)
    print(f't{i} = {t[i]}')
    print(f'y{i} = {y[i]}')

    alpha = t[i]
    beta = y[i]
    
```
