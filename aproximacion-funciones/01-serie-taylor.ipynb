# Serie de Taylor orden n

```python
from sympy import factorial, cos, plot
from sympy.abc import x
```

```python
n = 50
a = 1
```

```python
f = cos(x)*x**2 + 5 
f
```

```python
TAYLOR = 0
for n in range(n):
    TAYLOR += ((f.diff(x, n).subs(x, a) / factorial(n)) * (x - a)**n)

TAYLOR
```

```python
f_grafica = plot(f, show=False)
taylor_grafica = plot(TAYLOR, show=False)

f_grafica.append(taylor_grafica[0])

f_grafica.show()
```
